Now, finally combine all of them and define a function translate()

```python
def translate(seq):
    """
    Translate a string containing a nucleotide sequence into a string containing the corresponding sequence of amino acids.
    Nucleotides are translated in triplets using the table dictionary; each amino acid is encoded with a string of length 1.
    """
    # The explanation above is called 'docstring'
    table = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'_', 'TAG':'_',
    'TGC':'C', 'TGT':'C', 'TGA':'_', 'TGG':'W',
    }
    protein = ""
    if len(seq) % 3 == 0:
        for i in range(0,len(seq),3):
            codon = seq[i:i+3]
            protein += table[codon]
    return protein
```

### Counting Words

```python
def count_words(text):
    """
    Counts the number of times each word occurs in a given text.
    Returns a dictionary where keys are words, values are numbers.
    """
    word_counts = {}
    for word in text.split(" "): # opposite of join(), we used space because what separates words are spaces
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
    return word_counts

text = "Deneme deneme, deneme deneme."

count_words(text)
```

Notice that though we have only one word repeated 4 times in the example text, our function says there are 4 different words where each word repeated once. This is because our function separates capitalized letters and punctuations.

```python
def count_words_modified(text):
    text = text.lower()
    
    punctuations = [".", ",", ";", ":", "'", '"', "!", "?", "(", ")"]
    for punctuation in punctuations:
        text = text.replace(punctuation, "")    
    
    word_counts = {}
    for word in text.split(" "): # opposite of join(), we used space because what separates words are spaces
        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1
    return word_counts

count_words_modified(text)
```

### Introduction to k-Nearest-Neighbors (kNN) Classification
* The goal is to estimate or predict an output based on one or more inputs.
* If the response is quantitative such as a number that measures the height/weight, we call these problems **regression problems**.
* If the response is qualitative such as yes/no or blue/green, we call these problems **classification problems**.

In summary, when a new data point is added, we look its neighbors to determine its class

```python
import numpy as np

def distance(p1,p2):
    """
    Measures the distance between two given points p1 and p2.
    """
    sqr = np.power(p2 - p1,2) # np.power() takes the power of 2
    summation = np.sum(sqr) # np.sum() sums the elements in an array
    sqrt = np.sqrt(summation)
    return sqrt
```

### Majority Vote

We need to determine the most comman element in an array. For example, in [1,2,3,1,1,2] the most common vote or element is 1. We use a similar function to the function count_words we defined before

```python
def majority_vote(votes):
    vote_counts = {}
    for vote in votes:
        if vote in vote_counts:
            vote_counts[vote] += 1
        else:
            vote_counts[vote] = 1
    return vote_counts

votes = [1,2,3,1,2,3,1,2,3,3,3]
vote_counts = majority_vote(votes)
vote_counts
```

```python
import random

def majority_vote(votes):
    vote_counts = {}
    for vote in votes:
        if vote in vote_counts:
            vote_counts[vote] += 1
        else:
            vote_counts[vote] = 1
    winners = []
    for i in vote_counts.keys():
        if vote_counts[i] == max(vote_counts.values()):
            winners.append(i)
    return random.choice(winners) # if there are more than one winners, the code chooses one of them randomly

votes = [1,2,3,1,2,3,1,2,2,3,3] # there are same number of 3's and 2's
majority_vote(votes)
```

The function **find_nearest_neighbors()** down below returns the closes k points to p.

```python
def find_nearest_neighbors(p, points, k=5): # we choose k=5
    distances = np.zeros(points.shape[0])
    for i in range(len(distances)):
        distances[i] = distance(p, points[i])
    ind = np.argsort(distances)
    return ind[:k]


ind = find_nearest_neighbors(p, points, 2); print(points[ind])
```

If you check the image, you'll see the result is correct

```python
def knn_predict(p, points, outcomes, k=5): # 'outcomes' is the classes of the points
    ind = find_nearest_neighbors(p, points, k)
    return majority_vote(outcomes[ind]) # 'outcomes[ind]' returns not all elements but only those closes to the point p

outcomes = np.array([0,0,0,0,1,1,1,1,1])

knn_predict(np.array([2.5,2.7]),points,outcomes,k=2)
```

```python
def generate_synth_data(n=50):
    """
    Create two sets of points from bivariate normal distributions
    """
    points = np.concatenate((ss.norm(0,1).rvs((n,2)), ss.norm(1,1).rvs((n,2))), axis = 0)
    outcomes = np.concatenate((np.repeat(0,n),np.repeat(1,n)))
    return(points,outcomes)

n=20

(points,outcomes) = generate_synth_data(n)

plt.figure()
# points from the 1st category i.e. class 0
plt.plot(points[:n,0], points[:n,1], "ro");
# points from the 2n category i.e. class 1
plt.plot(points[n:,0], points[n:,1], "bo");
# plts.savefig("bivardata.pdf")

# '[:n,0]' first n rows and x coordinates are located in column 0
# '[:n,1]' first n rows and y coordinates are located in column 1
```

### Making a Prediction Grid
* make a prediction grid to examine some part of the predictor space
* how to use enumerate
* how to use Numpy meshgrid

```python
def make_prediction_grid(predictors, outcomes, limits, h, k):
    """Classify each point on the prediction grid."""
    (x_min,x_max,y_min,y_max) = limits
    xs = np.arange(x_min, x_max, h) # similar to np.linspace but only integer steps
    ys = np.arange(y_min, y_max, h)
    xx, yy = np.meshgrid(xs, ys)
    
    prediction_grid = np.zeros(xx.shape, dtype=int)
    # it's going to the same size as xx and yy, we define 'dtype=int' because we know the class clabel is either 0 or 1
    for i,x in enumerate(xs): # i loops in index, x loops in xs
        for j,y in enumerate(ys):
            p = np.array([x,y])
            prediction_grid[j,i] = knn_predict(p, predictors, outcomes, k)
    return (xx, yy, prediction_grid)
```

Numpy Meshgrid: **np.meshgrid()**
* takes in two or more coordinate vectors, say one vector containing the x values of interest and the other containing the y values of interest.
* returns matrices, the first containing the x values for each grid point and the second containing the y values for each grid point.

## Random Graphs
* how to write a function to build an Erdős-Rényi graph

* N is the number of nodes in the graph,
* p is the probability for any pair of nodes to be connected by an edge.
* Here's one way to think about it-- imagine starting with N nodes and no edges. You can then go through every possible pair of nodes and with probability p insert an edge between them. In other words, you're considering each pair of nodes once, independently of any other pair.You flip a coin to see if they're connected, and then you move on to the next pair.
* Although the NetworkX library includes an Erdos-Renyi graph generator, we'll be writing our own ER function to better understand the model.

```python
from scipy.stats import bernoulli

bernoulli.rvs(p=0.9) # 'p' defines the probability that we get 1
```

```python
bernoulli.rvs(p=0.1), bernoulli.rvs(p=0.5)
```

```python
a = [bernoulli.rvs(p=0.5) for i in range(10)]
a
```

```python
N = 20
p = 0.2

# create empty graph
# add all N nodes in the graph
# loop over all pairs of nodes
    # add an edge with prob p
    
G = nx.Graph()
G.add_nodes_from(range(N))

for node1 in G.nodes():
    for node2 in G.nodes(): # since we need a pair of nodes for an edg
        if bernoulli.rvs(p=p): #we define p=0.2
            G.add_edge(node1,node2)
# So we have adge only if bernoulli returns 1 with proability 0.2

nx.draw(G)
```

 Even though probability is low, the graph looks so dense because we consider the same pair twice. For example, consider the pair (1,10) we consider the situations node1=1,node2=10 and node1=10,node2=1.

 In order to get rid of this problem

```python
def er_graph(N,p):
    """Generate an ER graph"""
    G = nx.Graph()
    G.add_nodes_from(range(N))
    for node1 in G.nodes():
        for node2 in G.nodes():
            if bernoulli.rvs(p=p) and node1<node2:
                G.add_edge(node1,node2)
    return G

nx.draw(er_graph(20,0.2), node_size=40)
plt.savefig("er_graph.pdf")
```

* If p=1 is we have only 1 component since 'p=1' guarantees the existence of an edge between each node pair.
<br> <br>
* If p=0 is we have components as the same number of nodes we have since 'p=0' guarantees the absence of an edge between any node pair.

## Simple Linear Regression

$Y = \beta_0 + \beta_1X + \epsilon$

$\hat{y} = \hat{\beta}_0 + \hat{\beta}_1x$

$Y$: outcome
$\hat{y}$: prediction
$\hat{\beta}$s: estimates

$Y$ is a random variable, whereas $\hat{y}$ is a particular value.

$e_i = y_i - \hat{y}_i$

$e_i$: residual for ith observation

$RSS = {e_1}^2 + {e_2}^2 + ... + {e_n}^2$

$RSS $: residual sum of squares

```python
def compute_rss(y_estimate, y):
  return sum(np.power(y-y_estimate, 2))

def estimate_y(x, b_0, b_1):
  return b_0 + b_1 * x

rss = compute_rss(estimate_y(x, beta_0, beta_1), y)
rss
```

## Least Squares Estimation in Code
* compute the least squares estimate
* estimate parameter values that minimize the residual sum of squares (RSS) criterion

```python
# Suppose we know beta_0 but try to estimate beta_1

rss = [] # constains all rss values for different slopes
slopes = np.arange(-10, 15, 0.01)

for slope in slopes: 
    y_estimate = beta_0 + slope*x
    rss.append(np.sum((y - y_estimate)**2))
    
ind_min = np.argmin(rss)

print(ind_min) # returns the index location of the lowest value in rss

# Recall that we want to minimize rss so we need to select the lowest value

print("Estimate for the slope:", slopes[ind_min])

# Let's plot!

plt.figure()
plt.plot(slopes,rss)
plt.xlabel("Slope")
plt.ylabel("RSS");
```

**IMPORTANT:** I don't know whether you realize or not but we did an interesting job above. Knowing the fact that rss and slopes lists correspond to same index we first found the index that corresponds to the lowest rss. Then we print out the slop that corresponds that index.

## Assessing Model Accuracy
* to compute the mean squared error (MSE) to evaluate the performance of a regression model
* to separate data into training and test datasets
* why both underfitting and overfitting can be problematic

**To evaluate the performance of a regression model**: In the regression setting, the most commonly used measure is the **test mean squared error**, or MSE. Mean Squared Error (MSE) is given by an average over n data points.

$$MSE = \frac{1}{n}\sum_{i}^{n}\left(y_i-\hat{f}(x_i)\right)^2$$

$y_i$: observed outcome

$\hat{f}(x_i)$: prediction at the value $x_i$

**To evaluate how well a classifier performs**: we compute the **training error rate**, which is the proportion of errors the classifier makes when applied to training data. Analogously to the regression setting, we also have **test error rate**, which is the proportion of errors or misclassifications the classifier makes when applied to test data.

```python
from sklearn.model_selection import train_test_split
import numpy as np
import scipy.stats as ss

n = 500
beta_0 = 5
beta_1 = 2
beta_2 = -1

np.random.seed(1)

x_1 = 10 * ss.uniform.rvs(size = n)
x_2 = 10*ss.uniform.rvs(size = n)

X = np.stack([x_1,x_2], axis=1)
# not 'x' but 'X' because it is a matrix

y = beta_0 + beta_1*x_1 + beta_2*x_2 + ss.norm.rvs(loc=0, scale=1, size=n)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=1)
# 'train size': the propoertion of data that will be allocated to the training set
# 'random_state': spesifies the random seed
# primary motivation for splitting our model into training and testing data:
#evaluating how our model fits on unseen data, we can see how generalizable it is.

ln = LinearRegression(fit_intercept=True)

ln.fit(X_train, y_train)
# X_train: the matrix of predictors
# y_train: outcomes

# to test how well the model performs
ln.score(X_test, y_test)
```

Statistical models generally perform best when their capacity is appropriate for the complexity of the modeling task. Typically, more flexible models require estimating a greater number of parameters.

 Models that are too flexible can lead to overfitting, which means that the model starts to follow the noise in the data too closely. In the extreme case, the model can memorize the data points rather learn the structure of the data. The problem with this is that it generalizes poorly to unseen data.
 
 In contrast, if the model is too simple, it can underfit the data, in which case the model is not sufficiently flexible to learn the structure in the data.

 An extreme **overfitting** is simply just connection the dots together.
 
 An extreme **underfitting** is just a straight line.
